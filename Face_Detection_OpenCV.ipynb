```python
# 1️⃣ OpenCV installieren
!pip install opencv-python-headless numpy

# 2️⃣ Bibliotheken importieren
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Für Bildanzeige in Colab

# 3️⃣ Haarcascade-Modell laden (vortrainiertes Gesichtsmodell)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 4️⃣ Bild hochladen
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# 5️⃣ Bild einlesen & in Graustufen umwandeln
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 6️⃣ Gesichtserkennung
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# 7️⃣ Bounding Box um erkannte Gesichter zeichnen
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# 8️⃣ Ergebnis anzeigen
cv2_imshow(img)
```
